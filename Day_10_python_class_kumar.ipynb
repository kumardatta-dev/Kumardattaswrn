# Arguments

### Positional arguments

```python
def winner(name,prize):
    print(f"\tThe winner of the competion is {name.upper()}, he won a {prize}")
```

```python
winner('stuart','golden trophy')
```

```python
###---- in the above argument if we change the position of the name and prize the sentence is not formed correctly.
```

### Required arguments

```python
def display(c,d):
    print(c,d)
    display(c=10,d=20)
```

```python
display(10,20)
```
