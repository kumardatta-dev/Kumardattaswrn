# histogram ploting for only one numeric values 

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='darkgrid')

sns.histplot(data=DF,x='RATING',kde=True)
plt.show()
```

# Density chat for one numeric values 

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")

 
# Make default density plot
sns.kdeplot(DF['BRAND STAR'])
plt.show()
```

# BOX PLOT FOR

# A scatterplot displays the relationship between 2 numeric variables. For each data point, the value of its first variable is represented on the X axis, the second on the Y axis

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt

 
# use the function regplot to make a scatterplot
sns.regplot(x=DF["RATING"], y=DF["LITERS"])
 
# make a scatterplot without regression fit
#ax = sns.regplot(x=df["sepal_length"], y=df["sepal_width"], fit_reg=False)

plt.show()
```

```python
# library and dataset
import seaborn as sns
import matplotlib.pyplot as plt

 
# change shape of marker
sns.regplot(x=DF["RATING"], y=DF["PRICESS"], marker="+", fit_reg=False)

plt.show()
```

# BOX PLOT FOR THREE NUMERIC

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")


sns.boxplot(x=DF["RATING"], y=DF["PRICESS"], palette="Blues")
plt.show()
```

# one scatter and several numeric -- grouped scattert

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
 
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="Brand", y="PRICESS", data=DF, fit_reg=False, hue='RATING', legend=False )
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
plt. figure(figsize=(10000,10000))

plt.show()
```
