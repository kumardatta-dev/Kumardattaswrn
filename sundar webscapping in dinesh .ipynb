```python
TITLES=[]
RATINGS=[]
PRICES=[]
FEATURES=[]
for i in range(1,30):
    URL="https://www.flipkart.com/search?q=TVS&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(i)
    print(URL)
    PAGE=requests.get(URL)
    HTMLCODE=PAGE.text
    SOUP=BeautifulSoup(HTMLCODE)
     
    #TITLESS
    titles=SOUP.find_all("div",attrs={"class":"_4rR01T"})
    for i in titles:
        TITLES.append(i.text)
    
    #RATINGSS
    ratings=SOUP.find_all("div",attrs={"class":"_3LWZlK"})
    for i in ratings:
        RATINGS.append(i.text)
    
    #PRICESS
    prices=SOUP.find_all("div",attrs={"class":"_30jeq3 _1_WHN1"})
    for i in prices:
        PRICES.append(i.text)
    #FEATURESS
    features=SOUP.find_all("ul",attrs={"class":"_1xgFaf"})
    for i in features:
        FEATURES.append(i.text)

```

# IN CORRECT WAY OF DOING

```python
TITLES1=[]
PRICES1=[]
FEATURES1=[]
RATINGS1=[]
for i in range(1,20):
    URL="https://www.flipkart.com/home-kitchen/home-appliances/refrigerators/single-door~type/pr?sid=j9e%2Cabm%2Chzg&otracker=nmenu_sub_TVs+%26+Appliances_0_Single+Door&page={}".format(i)
    print(URL)
    PAGE=requests.get(URL)
    HTMLCODE=PAGE.text
    SOUP=BeautifulSoup(HTMLCODE)
    
    for x in SOUP.find_all("div",attrs={"class":"_2kHMtA"}):
        
        titles=x.find("div",attrs={"class":"_4rR01T"})
        if titles is None:
            TITLES1.append(np.NAN)
        else:
            TITLES1.append(titles.text)
            
        prices=x.find("div",attrs={"class":"_30jeq3 _1_WHN1"})
        if prices is None:
            PRICES1.append(np.NAN)
        else:
            PRICES1.append(prices.text)
        
        
        features=x.find("ul",attrs={"class":"_1xgFaf"})
        if features is None:
            FEATURES1.append(np.NAN)
        else:
            FEATURES1.append(features.text)
        
        rating=x.find("div",attrs={"class":"_3LWZlK"})
        if rating is None:
            RATINGS1.append(np.NAN)
        else:
            RATINGS1.append(rating.text)
        
        
    
```

```python
print(TITLES1[1])
```

# CREATING A DATA FRAME AND SAVE IT IN CSV FILE

```python
DF=pd.DataFrame({"PRODUCT":TITLES1,"RATING":RATINGS1,"PRICESS":PRICES1,"CONFIGURATION":FEATURES1})
```
